```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#Lectura de datos
r = pd.read_csv("metadata1_4.txt", sep="\t", engine = "python")
r
```

```python
df = pd.DataFrame({
    "station": r["station"],
    "kind": r["kind"]
})

df = df.groupby(["kind","station"]).size()
df = df.unstack()
ax = df.plot(kind="barh",figsize=(10,10), width=0.5, subplots=False)
ax.xaxis.grid(linestyle='--', linewidth=1)
ax.set_title("Histograma de Fonos por Clase y EstaciÃ³n", fontsize=20)
ax.set_ylabel("Clase", fontsize=20)
ax.set_xlabel("Cantidad", fontsize=20)
ax.invert_yaxis()

# plt.suptitle("Histograma de Fonos por Clase y EstaciÃ³n", fontsize=20)

# for i in range(len(ax)):   
#     ax[i].invert_yaxis()

```
